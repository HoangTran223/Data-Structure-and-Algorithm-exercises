```python
class Node:
    def __init__(self, data):
        self.right = None
        self.left = None
        self.data = data

    """Thêm phần tử"""
    def insert(self, data):     
    # Compare the new value with the parent node
        if self.data:                           # Nếu cha ko rỗng

            if data < self.data:               # data là con, self.data là cha
                if  self.left is None:          # Bên trái trống, thì bên trái dc add node
                    self.left = Node(data)
                else:
                    self.left.insert(data)      # Nếu không, thì nó add bên trái của cái data     
            
            elif data >= self.data:              # Đây là add nút bên phải
                if  self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data                    # Cha rỗng, nên add node vào
            
    """In cây"""
    def print_tree(self):
        if self.left:
            self.left.print_tree()
        print(self.data)
        if self.right:
            self.right.print_tree()
    
    """In-order Traversal: duyệt từ cây con trái -> gốc -> cây con phải"""
    def in_order_traversal(self, root):
      result = []
      if root:
         result = self.in_order_traversal(root.left)
         result.append(root.data)
         result += self.in_order_traversal(root.right)
      return result

    def pre_order_traversal(self, root):
        result = []
        if root:
            result.append(root.data)
            result += self.pre_order_traversal(root.left)
            result += self.pre_order_traversal(root.right)
        return result
        
root = Node(27)
root.insert(14)
root.insert(35)
root.insert(10)
root.insert(19)
root.insert(31)
root.insert(42)
root.insert(60)
root.insert(7)
root.insert(45)

print(root.pre_order_traversal(root))
print(root.in_order_traversal(root))    # Ta có thể sắp xếp nhờ duyệt BST theo thứ tự giữa    
```

SẮP XẾP CÁC NÚT THEO THỨ TỰ TRƯỚC

```python
"""Pre-order Traversal: duyệt từ nút gốc -> cây con trái -> cây con phải"""
def pre_order_traversal(self, root):
    result = []
    if root:
        result.append(root.data)
        result += self.pre_order_traversal(root.left)
        result += self.pre_order_traversal(root.right)
    return result
```

SẮP XẾP CÁC NÚT THEO THỨ TỰ SAU

```python
"""Post-order Traversal: duyệt từ cây con trái -> cây con phải -> gốc"""
def post_order_traversal(self, root):
    result = []
    if root:
        result = self.post_order_traversal(root.left)
        result += self.post_order_traversal(root.right)
        result.append(root.data)
    return result
```

Boundary Traversal (duyệt ranh giới của cây). Ta sẽ duyệt theo thứ tự:
+ Left boundary: đường đi từ gốc đến left-most node [nút(ko phải lá) mà ta đi max trái (ko đi dc mới chuyển phải) thu được]
+ Các nút lá: ta duyệt từ trái sang phải
+ Right boundary: ngược lại với trên

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.right = None
        self.left = None
        
    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            
            if data >= self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data

    def Left_Boundary(self, root):          # Hàm này ko in ra các lá, chỉ in nút
        if (root):
            if (root.left):
                print(root.data)            
                self.Left_Boundary(root.left)
            elif (root.right):
                print(root.data)
                self.Left_Boundary(root.right)

    def Right_Boundary(self, root):
        if (root):
            if (root.right):
                print(root.data)            
                self.Right_Boundary(root.right)
            elif (root.left):
                print(root.data)
                self.Right_Boundary(root.left)
    
    def Leaves(self, root):
        if (root):
            self.Leaves(root.left)               # Đệ qui để node đi đến cuối bên trái
            if root.left is None and root.right is None:
                print(root.data)
            self.Leaves(root.right)              # Đệ qui để node đi đến cuối bên phải

    def Boundary_Traversal(self, root):
        if (root):
            print(root.data)
            self.Left_Boundary(root.left)
            self.Leaves(root.left)
            self.Leaves(root.right)
            self.Right_Boundary(root.right)

root = Node(27)
root.insert(25)
root.insert(76)
root.insert(65)
root.insert(234)
root.insert(51)
root.insert(4567)
root.insert(174) 
root.insert(12)
root.insert(10)

root.Boundary_Traversal(root)          
```

Xóa 1 node trong Binary Search Tree (khác với xóa của Binary Tree):

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.right = None
        self.left = None

    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data >= self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data
            
    def in_order_traversal(self, root):
        result = []
        if root:
            result = self.in_order_traversal(root.left)
            result.append(root.data)
            result += self.in_order_traversal(root.right)
        return result

    def delete_node(self, root, key):
        if root is None:
            return root

        # Gọi đệ quy tổ tiên, sau bước này root sẽ là node cần xóa
        if root.data > key:
            root.left = self.delete_node( root.left, key)

        elif root.data < key:
            root.right = self.delete_node( root.right, key)  
        # Khi đã tìm thấy nút cần xóa:
        # Nếu 1 trong 2 con là rỗng:
        else:
            if root.left is None:
                temp = root.right
                del root
                return temp
            elif root.right is None:
                temp = root.left
                del root
                return temp
            else:
            # Nếu cả 2 con đều tồn tại:  tìm nút kế thừa(successor) có giá trị nhỏ nhất trong cây con bên phải.
            # sao chép giá trị của nó vào nút cần xóa, và sau đó xóa nút đó.
                successor_parent = root
                # Find successor:
                successor = root.right
                while successor.left is not None:    # Tức succ_parent sẽ thành cây nhánh phải, còn succ là nhánh trái con của succ_parent
                    successor_parent = successor            # Tiếp tục nó sẽ giống như  1
                    successor = successor.left              #                         /   \
                                                            #                              succ_pa
                                                            #                              / 
                                                            #                             succ / succ_pa ở lặp 2
                                                            #                             / 
                                                            #                             succ / succ_pa ở lặp 3
                                                            #                             / 
                                                            #                             succ (kết thúc lặp)

                # Đảm bảo cho việc sau khi xóa, BST vẫn giữ được tính chất:
                if successor_parent != root:                # Nếu successor không phải là con của node cần xóa
                    successor_parent.left = successor.right
                else:
                    successor_parent.right = successor.right
                
                # Sao chép giá trị successor vào nút cần xóa:
                root.data = successor.data

        return root

root = Node(27)
root.insert(25)
root.insert(76)
root.insert(65)
root.insert(234)
root.insert(51)
root.insert(4567)
root.insert(174) 
root.insert(12)
root.insert(10)

print(root.in_order_traversal(root))
root = root.delete_node(root, 65)

print(root.in_order_traversal(root))

root = root.delete_node(root, 27)
print(root.in_order_traversal(root))


```

Tìm predecessor và successor của 1 key trong BST:

```python
class Node():
    def __init__(self, data):
        self.data = data
        self.right = None
        self.left = None
        
    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data >= self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data
    
    def find_pre_suc(self, key):
        global find_pre, find_suc
        if self.data is None:
            return self.data
        if self.data == key:    # If key = root

            if self.left is not None: # Maximum value in left subtree is predecessor
                a = self.left
                while a.right:
                    a = a.right
                find_pre = a.data

            if self.right is not None: # Minimum value in right subtree is successor
                a = self.right
                while a.left:
                    a = a.left
                find_suc = a.data

        elif self.data > key:   # If key is smaller than root, go to left subtree
            find_suc = self.data
            self.left.find_pre_suc(key)
        else:                   # Higher, so go to right subtree
            find_pre = self.data
            self.right.find_pre_suc(key)
        

root = Node(27)
root.insert(25)
root.insert(76)
root.insert(65)
root.insert(234)
root.insert(51)
root.insert(4567)
root.insert(174) 
root.insert(12)
root.insert(10)

find_pre = None
find_suc = None
key = 4567

root.find_pre_suc(key)
 
if find_pre is not None:
    print ("Predecessor is", find_pre)
else:
    print ("No Predecessor")
 
if find_suc is not None:
    print ("Successor is", find_suc)
else:
    print ("No Successor")
 
            
```

Thêm phần tử trên cây AVL:

```python
class AVLNode:
    def __init__(self, value):
        self.value = value
        self.right = None
        self.left = None
        self.height = 1
    
    def preorder_traversal(self, root):
        a = []
        if root:
            a.append(root.value)
            a += self.preorder_traversal(root.left)
            a += self.preorder_traversal(root.right)
        return a
    
    def insert_AVL(self, root, key):
        if not root:
            return AVLNode(key)
        elif key < root.value:
            root.left = self.insert_AVL(root.left, key)
        else:
            root.right = self.insert_AVL(root.right, key)

        # Update the height of the ancestor node
        root.height = 1 + max(self.getHeight(root.left), self.getHeight(root.right)) 
  
        # Balance factor
        balance = self.getBalance(root) 
  
        # If the node is unbalanced
        if balance > 1 and key < root.left.value:       # Case 1
            return self.rightRotate(root) 

        elif balance < -1 and key > root.right.value:   # Case 2
            return self.leftRotate(root)

        elif balance > 1 and key > root.left.value:     # Case 3
            root.left = self.leftRotate(root.left)
            return self.rightRotate(root)

        elif balance < -1 and key < root.right.value:   # Case 4 
            root.right = self.rightRotate(root.right)
            return self.leftRotate(root)
        
        return root

    def rightRotate(self, z):                           # Follow the diagram : z
        y = z.left                                      #                    /        
        t = y.right                                     #                   y  
        # Rotating:                                     #                    \
        y.right = z                                     #                      t
        z.left = t
                                      
        z.height = 1 + max(self.getHeight(z.left),      # Bắt buộc phải cập nhật z trước, do z là con của y
                        self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left),
                        self.getHeight(y.right))
        # Return the new root
        return y
    
    def leftRotate(self, z):                            # Follow the diagram : z
        y = z.right                                     #                       \
        t = y.left                                      #                        y
        # Rotating:                                     #                       / 
        y.left = z                                      #                      t
        z.right = t 

        z.height = 1 + max((self.getHeight(z.left), self.getHeight(z.right)))       
        y.height = 1 + max((self.getHeight(y.left), self.getHeight(y.right)))
        # Return the new root
        return y 
    
    def getHeight(self, root):
        if not root:
            return 0
        else:
            return root.height
    
    def getBalance(self, root):
        if not root:
            return 0
        else:
            return self.getHeight(root.left) - self.getHeight(root.right)


# root = AVLNode(13)
# root = root.insert_AVL(root, 10)
# root = root.insert_AVL(root, 15)
# root = root.insert_AVL(root, 5)
# root = root.insert_AVL(root, 11)
# root = root.insert_AVL(root, 16)
# root = root.insert_AVL(root, 4)
# root = root.insert_AVL(root, 6)
# root = root.insert_AVL(root, 7)
root = AVLNode(5)
root = root.insert_AVL(root, 7)
root = root.insert_AVL(root, 2)
root = root.insert_AVL(root, 1)
root = root.insert_AVL(root, 6)
root = root.insert_AVL(root, 4)
root = root.insert_AVL(root, 9)
root = root.insert_AVL(root, 3)
root = root.insert_AVL(root, 16)
root = root.insert_AVL(root, 15)

print(root.preorder_traversal(root))
```

```python
class AVLNode():
    def __init__(self, value):
        self.value = value
        self.right = None
        self.left = None
        self.height = 1
    
    def preorder_traversal(self, root):
        a = []
        if root:
            a.append(root.value)
            a += self.preorder_traversal(root.left)
            a += self.preorder_traversal(root.right)
        return a
    
    def insert_AVL(self, root, key):
        if not root:
            return AVLNode(key)
        elif key < root.value:
            root.left = self.insert_AVL(root.left, key)
        else:
            root.right = self.insert_AVL(root.right, key)

        # Update the height of the ancestor node
        root.height = 1 + max(self.getHeight(root.left), self.getHeight(root.right)) 
  
        # Balance factor
        balance = self.getBalance(root) 
  
        # If the node is unbalanced
        if balance > 1 and key < root.left.value:       # Case 1
            return self.rightRotate(root) 

        elif balance < -1 and key > root.right.value:   # Case 2
            return self.leftRotate(root)

        elif balance > 1 and key > root.left.value:     # Case 3
            root.left = self.leftRotate(root.left)
            return self.rightRotate(root)

        elif balance < -1 and key < root.right.value:   # Case 4 
            root.right = self.rightRotate(root.right)
            return self.leftRotate(root)
        
        return root

    def rightRotate(self, z):                           # Follow the diagram : z
        y = z.left                                      #                    /        
        t = y.right                                     #                   y  
        # Rotating:                                     #                    \
        y.right = z                                     #                      t
        z.left = t
                                      
        z.height = 1 + max(self.getHeight(z.left),      
                        self.getHeight(z.right))
        y.height = 1 + max(self.getHeight(y.left),
                        self.getHeight(y.right))
        # Return the new root
        return y
    
    def leftRotate(self, z):                            # Follow the diagram : z
        y = z.right                                     #                       \
        t = y.left                                      #                        y
        # Rotating:                                     #                       / 
        y.left = z                                      #                      x
        z.right = t 

        z.height = 1 + max((self.getHeight(z.left), self.getHeight(z.right)))
        y.height = 1 + max((self.getHeight(y.left), self.getHeight(y.right)))
        # Return the new root
        return y 
    
    def getHeight(self, root):
        if not root:
            return 0
        else:
            return root.height
    
    def getBalance(self, root):
        if not root:
            return 0
        else:
            return self.getHeight(root.left) - self.getHeight(root.right)

    def delete_node(self, root, key):
        # Standard BST:
        if root is None:
            return root
        elif key < root.value:
            root.left = self.delete_node(root.left, key)
        elif key > root.value:
            root.right = self.delete_node(root.right, key)
        else:
            if root.left is None:
                temp = root.right
                root = None
                return temp
            elif root.right is None:
                temp = root.left
                root = None
                return temp
    
            # Tối ưu hơn code phiên bản đầu
            temp = self.min_node(root.right)        # GTNN của cây con của nút phải
            root.value = temp.value
            root.right = self.delete_node(root.right, temp.value)
 
        # If the tree has only 1 node
        if root is None:
            return root
 
        # Thực hiện tương tự thao tác chèn
        root.height = 1 + max(self.getHeight(root.left),
                            self.getHeight(root.right))

        balance = self.getBalance(root)
 
        if balance > 1 and self.getBalance(root.left) >= 0:         # Case 1: Left - Left, ko dùng như cách chèn, do ta phải duyệt đến tận gốc, và theo code thì key đã = root.value 
            return self.rightRotate(root)                           # Nếu để giống hệt chèn, thì nó sẽ báo lỗi rất nhiều

        if balance < -1 and self.getBalance(root.right) <= 0:       # Case 2: Right - Right
            return self.leftRotate(root)
 
        if balance > 1 and self.getBalance(root.left) < 0:
            root.left = self.leftRotate(root.left)
            return self.rightRotate(root)

        if balance < -1 and self.getBalance(root.right) > 0:
            root.right = self.rightRotate(root.right)
            return self.leftRotate(root)
 
        return root
 
    def min_node(self, root):                   
        if root is None or root.left is None:
            return root
        return self.min_node(root.left)


root = AVLNode(9)
nodes =  [5, 10, 25, 6, 11, 7, 1, 2]
for node in nodes:
    root = root.insert_AVL(root, node)
print(root.preorder_traversal(root))
key = 7
root = root.delete_node(root, key)

print(root.preorder_traversal(root))
```
